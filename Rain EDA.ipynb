```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('POWER_SinglePoint_Daily_2000-2021.txt', header= 12, delimiter= '\s+')
```

```python
df
```

```python
df.info()
```

```python
df['DATE'] = df['DATE']+2000000
```

```python
df
```

```python
df['DATE'] = pd.to_datetime(df['DATE'], format='%Y%j').dt.strftime('%Y-%m-%d')
```

```python
df
```

```python
df = df.set_index('DATE')
```

```python
dt = pd.date_range('2000-01-01', '2021-03-31')
idx = pd.DatetimeIndex(dt)
df = df.reindex(idx)
```

```python
tahun_akhir = df['2016':'2021']
tahun_akhir['day_of_month'] = tahun_akhir.index.day
tahun_akhir['day_of_week'] = tahun_akhir.index.dayofweek
tahun_akhir['month'] = tahun_akhir.index.month
tahun_akhir['year'] = tahun_akhir.index.year
```

```python
fig,(ax1, ax2, ax3, ax4)= plt.subplots(nrows=4)
fig.set_size_inches(18, 28)

sns.pointplot(data=tahun_akhir, x='day_of_month', y='RAIN', hue = 'year', ax=ax1)
sns.pointplot(data=tahun_akhir, x='day_of_week', y='RAIN', hue='year', ax=ax2)
sns.pointplot(data=tahun_akhir, x='month', y='RAIN', hue='year', ax=ax3)
sns.pointplot(data=tahun_akhir, x='year', y='RAIN', ax=ax4)
```
